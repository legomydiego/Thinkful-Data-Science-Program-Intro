```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exponential = np.random.exponential(1/50,50)
exp_mean = exponential.mean()
exp_std = exponential.std()
```

```python
plt.hist(exponential,bins=50)
plt.axvline(exp_mean, color='r', linestyle='solid', linewidth=2)
plt.axvline(exp_mean+exp_std, color='y', linestyle='dashed', linewidth=2)
plt.axvline(exp_mean-exp_std, color='y', linestyle='dashed', linewidth=2)
plt.show()
```

```python
normal1 = np.random.normal(5,0.5,5000000)
```

```python
plt.hist(sum_normal,bins=50)
plt.axvline(sum_normal.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(sum_normal.mean()+sum_normal.std(), color='y', linestyle='dashed', linewidth=2)
plt.axvline(sum_normal.mean()-sum_normal.std(), color='y', linestyle='dashed', linewidth=2)
plt.show()
```
