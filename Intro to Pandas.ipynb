```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('UNdata_Export_20181111_213008506.csv')
```

```python
df.head(20)
```

```python
df.describe()
```

```python
df[df['Country or Area']=='United States']
```

```python
df[df['Country or Area']=='United States'].plot(x='Year',y='Value')
```

```python
USA = df[df['Country or Area']=='United States']
```

```python
USA.head()
```

```python
USA.sort_index(ascending=False,inplace=True)
```

```python
plt.figure(figsize=(15,5))
plt.scatter(x=USA['Year'],y=USA['Value'])
plt.tight_layout()
plt.show()
```

```python
DR = df[df['Country or Area']=='Dominican Republic']
```

```python
DR.sort_index(ascending=False,inplace=True)
```

```python
plt.figure(figsize=(15,5))
plt.scatter(x=DR['Year'],y=DR['Value'])
plt.tight_layout()
plt.show()
```

Both graphs are similar, but the USA chart's scale is 100 times larger. Both countries experienced a slowdown with the financial crisis in 2008, but have recovered nicely
